# Background
Contemplative practice is an umbrella term which refers to daily activities that sustain a healthy, interconnected mind-body-spirit. These practices include gentle movement, stretching, breathing exercises, mindfulness meditation, and expressing gratitude. Engaging in contemplative practice brings many benefits, including slowing the process of aging in the brain (Lazar et al., 2005), strengthening the immune system (Davidson et al., 2003), and regulating stress hormones (Tang et al., 2007). Thus, interventions which promote engagement in contemplative practice are warranted. Unfortunately, the National Health Interview Survey has found that men, people of lower socioeconomic status, and people of color are less likely to engage in these practices (Kachan et al., 2017; Olano et al., 2015). However, it is possible that these groups have less access to the proper guidance and knowledge about contemplative practice. This study seeks to evaluate whether there are certain groups less likely to engage in contemplative practice even after they have been provided the necessary guidance.
    
If demographic differences are observable, the current study also hopes to be able to make recommendations for interventions that promote engagement in contemplative practice from wide audiences. A developed and widely used theoretical basis for designing interventions that change people’s behaviors is the Theory of Planned Behavior (Fishbein & Ajzen, 2010). The Theory of Planned Behavior postulates that attitudes, perceived social norms, and self-efficacy around engaging in a particular behavior predict intentions to engage in the behavior, which in turn influences actual engagement in the behavior. Attitudes, norms and self-efficacy factors that prove to be particularly important in influencing intentions can then be targeted by interventions aiming to increase engagement in the ideal behavior. Specifically, the attitudes, norms and self-efficacy which differentiate the most between demographic groups can be targeted in order to engage a more diverse audience in a set of behaviors. Thus, the Theory of Planned Behavior was used in the current study to provide suggestions on how engagement in contemplative practice can spread across a wider population.

## Study Protocol
Participation in the study was advertised as a voluntary, uncompensated opportunity to be guided through steps to quiet the mind, relax and appreciate the good in life. Potential participants were contacted through email lists and Facebook advertisements. Email lists used included Stanford’s wellness initiatives, which reach both Stanford employees and students. When participants signed up for the study and provided their demographic information, they selected a starting date from which they planned to start a week of activities that sustain a healthy, interconnected Mind-Body-Spirit. On this starting date, they received a survey by text or email, depending on the participant’s preference, with nine measures assessing their attitudes, perceived norms, self-efficacy and intentions to engage in contemplative practice for at least five minutes, everyday, for the next week. After completing this survey, participants were provided with resources explaining the benefits of contemplative practice as well as links to seven guided contemplative practices. Subjects were instructed to engage in at least five minutes of guided contemplative practice each day for a week. Their engagement in these practices was measured though self-reports they received each day of the week that asked them to report which guided contemplative practices they engaged in on the previous day. On the last day of the week, participants were also prompted to report on how many days over the last week they engaged in each of the seven guided contemplative practices. Lastly, each participant’s actual behavior was tracked by measuring the dates and times when a participant opened and closed the links to each practice.

#### Summed Self-Reported Engagement
The 141 participants with some indication of having engaged in contemplative practice during the study were assigned a second behavioral score, the summed self-reported engagement. This score attempts to capture how often these participants engaged in the practices. During preprocessing, this score was calculated using both the daily self-reports participants filled out each day during their week of behavior tracking, as well as the final self-report sent out on the last day of the week. Data from the behavioral tracking was not incorporated in this later score as data from when participants opened and closed the links was not well correlated with the self-reports and did not appear to reliably capture every time a participant engaged in a practice. The summed self-reported engagement was calculated using the maximum between the summed final self-report and the sum of all daily self-reports. The maximum was taken between the daily self-reports and final self-report to account for participants who did not complete every self-report.

This score is stored as 'behavior_continuous_normal'

### Defining Functions to Make Cross Validation Splits and Train and Test Logistic Regression Models

```python
#Importing logistic regression packages
from sklearn.linear_model import (LogisticRegression, LogisticRegressionCV)
```

```python
#I will eventually want to take a look at the confusion matrices for these logistic models, so I define a function below
#to help me visualize them
def visualize_confmat(cmat, ucats, fig=None):
    """
        Visualizes confusion matrix cmat with unique category labels ucats
    """
    #Graphing confusion matrix
    if fig is None:
        plt.matshow(cmat,cmap=plt.cm.Blues,vmin=0,vmax=1)
    else:
        fig.gca().matshow(cmat)
    #Adding in actual percentages to each cell
    for i in range(cmat.shape[0]):
        for j in range(cmat.shape[1]):
            plt.text(j-.2, i, '%.2f%%' % (cmat[i, j] * 100), color='k',fontsize=14)
    #Adding labels and ticks to matrix display
    M = len(ucats)
    plt.xticks(np.arange(M), ucats,fontsize=18)
    plt.yticks(np.arange(M), ucats,fontsize=18)
    plt.xlabel('Predicted',fontsize=22)
    plt.ylabel('Actual',fontsize=22)
```

```python
#To calculate the confusion matrix, I will be using code from sklearn
from sklearn.metrics import confusion_matrix
```

```python
#Defining a function to split the data while making sure that the labels I will be predicting with the logistic regression
#models will be balanced. This function returns splits of the indices in targets.
def get_splits_by_class(targets,indices,num_splits=30,
                        percent4test=.5):
    """
    constructing splits for cross validation
    
    arguments:
        num_splits - how many cross validation splits to return
        percent4test - what percent of indicies in targets should be allocated for testing, default does half splits
        targets - vector, the variable I am trying to predict, which contains categories I hope to balance in the splits
        indices - potential indices for splits, must only index participants with complete data for the variables I am
                  using
    """
    
    err_msg = 'Please indicate percent as decimal from 0 to 1'
    assert percent4test < 1, err_msg
    
    #Figuring out how many classes I have and what is the smallest sample size for a class
    classes = np.unique(targets[indices]) #All of the classes in targets
    smallest_sample = len(targets) #Will store the size of the class with the smallest sample
    #Looping across classes to see which one has the smallest sample size
    for cls in classes:
        if np.sum(targets[indices] == cls) < smallest_sample:
            smallest_sample = np.sum(targets[indices] == cls)
    
    #Calculating the number from each class to use for testing
    num_testPerSample = int(percent4test*smallest_sample) #By using int, it will round down
    
    #Construct splits
    splits = []
    #For each split I want
    for _split_ind in range(num_splits):
        np.random.seed(_split_ind) #Initializing seed
        #Storing testing and training indices in lists
        test_indices = []
        train_indices = []
        for cls in classes:
            indices4class = indices[(targets[indices] == cls)[:,0]] #Indices for this class
            #Randomly choosing num_testPerSample indices from indices for this split
            test_indices4class = np.random.choice(indices4class,num_testPerSample,False) #False indicates drawing without
                                                                                         #replacement
            #Remaining indices could possibly be used for training, don't want to overlap training and testing
            possible_train_indices4class = np.unique(list(set(indices4class).difference(test_indices4class)))
            #I want the same number in training from each class, which is smallest_sample - num_testPerSample
            train_indices4class = np.random.choice(possible_train_indices4class,
                                                   smallest_sample-num_testPerSample,False)
            #Adding indices to testing and training arrays
            test_indices = np.append(test_indices,test_indices4class)
            train_indices = np.append(train_indices,train_indices4class)
        #Sorting Indices to make sure they can be used to index in data
        test_indices = np.sort(test_indices).astype(int)
        train_indices = np.sort(train_indices).astype(int)
        #Storing Indices
        split = {'train': train_indices, 'test': test_indices}
        splits.append(split)
        
    return splits
```

```python
#I'll need a function that trains and tests the logistic regressors
def train_and_test_logregressor(feature,
                                label,
                                splits,
                                complete,
                                model_class=LogisticRegression,
                                model_args={},
                                return_models=False):
    """Training and testing classifiers
    
        Arguments:
            features - predictor array organized as number of participants by number of features
            labels - vector of labels to be predicted
            splits - split indices of data
            complete - indices for complete data for the features and labels I am using
            model_class - what kind of classifer, default is LogisticRegression
            model_args - arguments to initialize the classifer
            return_models - whether or not I want the function to return each of the trained models
            
        Returns:
            Dictionary containing
                test and train confusion matrices = array of confusion matrices from each split on testing and training data
                mean_confusion_matrix for testing and training = average of all confusion matrices across all splits on 
                                                                 testing data
                std_confusion_matrix for testing and training = standard deviation for each cell in the confusion matrix 
                                                                across splits on testing data
                mean and standard deviation of accuracies for testing and training
                models - if return_models is true, array of all models trained
    """
    
    train_confmats = []
    test_confmats = []
    train_accuracies = []
    test_accuracies = []
    clfs = [] #stores classifiers
    
    #Using cross validation to find best C for the feactures and labels
    cvlog = LogisticRegressionCV()
    cvlog.fit(feature[complete],label[complete])
    c = cvlog.C_[0]
    
    #Figuring out the classes I am predicting
    classes = cvlog.classes_
    
    #Looping across splits
    for split in splits:
        
        #instantiate new classifier
        clf = model_class(C=c)
        
        #organizing data into training and testing for this split
        train_inds = split['train']
        test_inds = split['test']
        train_features = feature[train_inds]
        train_labels = label[train_inds]
        test_features = feature[test_inds]
        test_labels = label[test_inds]
        
        #fitting model
        clf.fit(train_features,train_labels)
        
        #Getting confusion matrices for training and testing
        train_predictions = clf.predict(train_features)
        train_confmat = confusion_matrix(train_labels,train_predictions)
        
        test_predictions = clf.predict(test_features)
        test_confmat = confusion_matrix(test_labels,test_predictions)
        
        #Converting confusion matrix to percents
        train_confmat = train_confmat.astype('float') / train_confmat.sum(axis=1)[:,np.newaxis]
        test_confmat = test_confmat.astype('float') / test_confmat.sum(axis=1)[:,np.newaxis]
        
        #Calculating mean accuracy on testing and training data
        train_accuracy = clf.score(train_features,train_labels)
        test_accuracy = clf.score(test_features,test_labels)
        
        #Storing results from this split
        train_confmats.append(train_confmat)
        test_confmats.append(test_confmat)
        train_accuracies.append(train_accuracy)
        test_accuracies.append(test_accuracy)
        
        if return_models:
            clfs.append(clf)
    
    #Calculating mean and standard deviation of the accuracies across splits
    mean_train_confmat = np.mean(train_confmats,axis=0)
    std_train_confmat = np.std(train_confmats,axis=0)
    mean_test_confmat = np.mean(test_confmats,axis=0)
    std_test_confmat = np.std(test_confmats,axis=0)
    mean_train_accuracy = np.mean(train_accuracies)
    std_train_accuracy = np.std(train_accuracies)
    mean_test_accuracy = np.mean(test_accuracies)
    std_test_accuracy = np.std(test_accuracies)
    
    #Organizing results into a dictionary
    results = {'train_confusion_matrices': train_confmats,
               'test_confusion_matrices': test_confmats,
               'mean_train_confusion_matrix': mean_train_confmat,
               'std_train_confusion_matrix': std_train_confmat,
               'mean_test_confusion_matrix': mean_test_confmat,
               'std_test_confusion_matrix': std_test_confmat,
               'mean_train_accuracy': mean_train_accuracy,
               'std_train_accuracy': std_train_accuracy,
               'mean_test_accuracy': mean_test_accuracy,
               'std_test_accuracy': std_test_accuracy,
               'train_accuracies': train_accuracies,
               'test_accuracies': test_accuracies,
               'classifiers': clfs}
    
    return results
```

The way I have defined my train_and_test_logregressor function above requires that my splits do not index incomplete data. Thus, I need to calcuate all of the indices that contain complete data for the features and labels I am using.

```python
def complete_indices(variables,returnBool=False):
    """
        Finds which indices in the dataset have complete data for all of the variables
        specified.
        
        Arguments - list of variable names in data, and an indication of whether the returned index should be boolean
        Returns - if returnBool is false, integer array of the indices with complete data; otherwise boolean indexing array
    """
    
    #I want to store boolean index vectors that show for each variable what indices have complete data in a list.
    completeByVar = []
    #To do this, I'll run a for loop over variables and append each vector to completeByVar
    for var in variables:
        completeByVar.append(~np.isnan(data[var]))
    
    #Boolean index for indices in data with complete data for all variables
    completeBool = np.all(completeByVar,axis=0)
    
    #Converting boolean index to integer index
    all_indices = np.array(range(len(completeBool)))
    complete_indx = all_indices[completeBool[:,0]]
    
    if returnBool:
        return completeBool
    else:
        return complete_indx
```

To make things easier with using logistic regression analysis, I defined the function below that gets complete indices, splits the indices, and trains and tests the regressors.

```python
def logregressor_analysis(feature_variables,label_variable,splits=None):
    """
        Finds indices containing complete data in data for all of the variables listed (feature_variables and 
        label_variables) in order to calculate splits and train and test sklearn log regressors unless I already have splits
        
        arguments:
            feature_variables - list of the dictionary indices in data for the features I would like to use
            label_variable - list of dictionary index in data for the label I would like to predict
            splits - if I want to use the same splits as a previous analysis, I can put them here
            
        returns results from train_and_test_logregressor and the splits used
    """
    #Combining all variable names
    all_variables = feature_variables + label_variable
    #Calculating indices with complete data for all of these variables
    completeIndx = complete_indices(all_variables)
    if splits is None:
        #Getting splits of these indices
        splits = get_splits_by_class(data[label_variable[0]],completeIndx)
    #Storing list of feature data
    features = []
    for feature in feature_variables:
        features.append(data[feature])
    #Converting into a matrix of all predicing data
    features = np.array(features)[:,:,0].T
    #Returning results from training and testing across the splits
    return train_and_test_logregressor(features,data[label_variable[0]],splits,completeIndx), splits
```

I won't use cross validation splits here as my models aren't fitting training data well.

```python
#Importing ridge regression package
from sklearn.linear_model import RidgeCV
```

```python
#I'll define a function that fits the linear regression and reports the r-squared statistic
def fit_linear_regression(feature_variables,target_variables,condition=None):
    """
        Returns the r-squared statistic of a linear regression model fit on all of the available data for a given condition
        
        Arguments:
            feature_variables - list of variable keys in data that I want to use as features
            target_variables - list of variable keys in data that I want to use a labels
            condition - list of boolean indices, for instance showing which participants have at least a bachelor's degree
    """
    #Calculating what indices of data to use for this linear regression
    #Boolean indices for complete data
    complete_data = [complete_indices(feature_variables + target_variables,returnBool=True)]
    #Combined boolean indices for complete data and condition
    if condition is None:
        boolIndx = complete_data
    else:
        boolIndx = np.all(complete_data+condition,axis=0)
    #Making arrays of features and targets
    features = []
    for feature in feature_variables:
        features.append(data[feature][boolIndx])
    features = np.array(features).T
    targets = []
    for target in target_variables:
        targets.append(data[target][boolIndx])
    targets = np.array(targets).T
    #Instantiating linear model
    lm = RidgeCV()
    #Fitting model, not using training/testing splits
    lm.fit(features,targets)
    #Returning R-squared coefficient
    return lm.score(features,targets)
```

## Attitudes, Norms and Self Efficacy for Participants of Similar Intentions and Behavior
My next major goal with these analyses is to see whether attitudes, norms and self-efficacy of participants with the same intentions and level of engagement were similar. 
### Representational Similarity Matrices
To do this, I'll start off by plotting matrices similar to the representational similarity matrices, to see if participants of the same intentions and level of engagement had more correlated attitudes, norms and self-efficacies. Essentially, these matrices compare each participant's attitudes, norms and self-efficacies with those of all other participants, and thus take the same of number of participants by number of participants.

```python
def rsm_like(labels,labelNames,items2correlate=ANS_vars):
    """
        Calculates a pearson correlation matrix comparing attitudes, norms and self-efficacies between subjects and displays
        
        Arguments:
            items2correlate - variables in data which I want to compare between subjects
            labels - variable in data along which I want to subdivide and sort participants
            labelNames - the actual names for each label (e.g. 'Under 25')
            
        Returns:
            actualMat - actuall rsm like matrix
            plots ideal matrix and actual matrix
    """
    #Calculating the number of unique labels
    num_labels = np.unique(data[labels][~np.isnan(data[labels])])
    #Storing data from items2correlate into a matrix
    ANS_data = []
    for item in items2correlate:
        ANS_data.append(data[item])
    ANS_data = np.array(ANS_data)[:,:,0].T #Converting into a matrix from a list
    #Using only complete data
    complete = complete_indices(ANS_vars)
    ANS_data = ANS_data[complete]
    
    #Sorting ANS_data by labels
    dataByLabel = []
    #Calculating number of participants per label
    numPerLabel = []
    for label in num_labels:
        data4label = ANS_data[(data[labels][complete] == label)[:,0],:]
        dataByLabel.append(data4label)
        numPerLabel.append(data4label.shape[0])
    
    #Stacking dataByLabel into a matrix taking the shape of number of participants by number of variables being correlated
    sortedData = dataByLabel[0]
    for i in range(1,len(dataByLabel)):
        sortedData = np.vstack((sortedData,dataByLabel[i]))
    
    #Calculating actual between subject correlations
    actualMat = np.corrcoef(sortedData)
    
    #Calculating an ideal matrix using synthetic data, starting off by assigning low random correlations normally 
    #distributed around 4 to all between-subject comparisons
    idealMat = np.random.normal(loc=0.4,scale=0.4,size=actualMat.shape)
    
    #For each area in idealMat where participants are from the same label, I want to increase their synthetic correlation
    j = 0
    k = 0
    #Storing what will be the tick marks used in making figures
    ticks = [0]
    while j < np.sum(numPerLabel):
        h = j
        j += numPerLabel[k]
        idealMat[h:j,h:j] = np.random.normal(loc=0.65,scale=0.29,size=(numPerLabel[k],numPerLabel[k]))
        k += 1
        ticks.append(j)
        
    #The diagonal in idealMat should be 1
    diagonal = np.diag_indices(idealMat.shape[0])
    idealMat[diagonal] = 1
    
    #Plotting matrices
    plt.subplots(1,2,figsize=(14,14))
    #The left plot will have the ideal matrix
    plt.subplot(1,2,1)
    plt.imshow(idealMat,cmap='inferno',vmin=-1,vmax=1)
    plt.xticks(ticks,labelNames,fontsize=14)
    plt.yticks(ticks,labelNames,fontsize=14)
    plt.xlabel('Sorted Survey Responses\nfor Each Participant',fontsize=18)
    plt.ylabel('Sorted Survey Responses\nfor Each Participant',fontsize=18)
    plt.title('Ideal RSM-Like Matrix',fontsize=22)
    #The right plot will have the actual matrix
    plt.subplot(1,2,2)
    plt.imshow(actualMat,cmap='inferno',vmin=-1,vmax=1)
    plt.xticks(ticks,labelNames,fontsize=14)
    plt.yticks(ticks,labelNames,fontsize=14)
    plt.xlabel('Sorted Survey Responses\nfor Each Participant',fontsize=18)
    plt.ylabel('Sorted Survey Responses\nfor Each Participant',fontsize=18)
    plt.title('Actual RSM-like Matrix',fontsize=22)
    
    plt.subplots_adjust(wspace=.6)
    
    return actualMat
```

### Split-Half Reliability
I now want to see if the attitudes, norms and self-efficacies of participants with similar intentions and levels of engagement are reliably similar. To do this, I'll divide participants of each level of intention and engagement into multiple half splits and correlate the mean attitudes, norms and self-efficacies of each half with the other. I'll average these correlations across splits and see if the attitudes, norms and self-efficacies of participants from the same level of intention and engagement are reliably correlated.

```python
#Defining a function for this split-half analysis
def split_half_analysis(label_var,items2correlate=ANS_vars):
    """
        Performs the split half analysis described above
        
        Arguments:
            label_var - the variable in data (intentions or behavior) of interest
            items2correlate - the survey items that will be averaged across each half of participants from the same group
    """
    #Getting unique labels from label_var
    complete4label = complete_indices([label_var])
    labels = np.unique(data[label_var][complete4label])
    
    #Calculating the number of splits that I need, which is 10 times the size of the largest label
    largest_label = 0
    for label in labels:
        if largest_label < np.sum(data[label_var] == label):
            largest_label = np.sum(data[label_var] == label)
    num_splits_needed = 10*largest_label
    
    #Calculating complete indices for all variables used
    completeAllVar = complete_indices([label_var]+items2correlate)
    
    #Getting splits
    splits = get_splits_by_class(data[label_var],completeAllVar,num_splits=num_splits_needed)
    
    #Storing all of the data from items2correlate into a matrix
    itemsMat = []
    for item in items2correlate:
        itemsMat.append(data[item])
    itemsMat = np.array(itemsMat)[:,:,0].T
    
    corrMats = []
    for split in splits:
        #Instantiating a matrix that will contain all of the split-half correlations for this split
        corrMat = np.zeros((len(labels),len(labels)))
        #Looping over labels to fill in matrix values
        for half1label in range(len(labels)):
            for half2label in range(len(labels)):
                #Finding the indices corresponding to the appropriate labels in the first and second half
                half1labelIndx = data[label_var][split['train']] == half1label
                half2labelIndx = data[label_var][split['test']] == half2label
                #Storing the data for the appropriate labels in these halves
                half1labelData = itemsMat[split['train'][half1labelIndx[:,0]]]
                half2labelData = itemsMat[split['test'][half2labelIndx[:,0]]]
                #Averaging data from each half
                avgHalf1LabelData = np.mean(half1labelData,axis=0)
                avgHalf2LabelData = np.mean(half2labelData,axis=0)
                #Correlating the two averages
                r = np.corrcoef(avgHalf1LabelData,avgHalf2LabelData)[1,0]
                #Storing in corrMat
                corrMat[half1label,half2label] = r
        corrMats.append(corrMat)
        
    #Averaging across splits
    avgCorrMat = np.mean(corrMats,axis=0)
    
    return avgCorrMat
```

```python
#Function to calculate and show split half matrix
def show_split_half(label_var,categoryNames):
    """
        Displays split half correlation matrix
        
        Arguments:
            label_var - across which label I want to calculate splits and correlated
            categoryNames - names of categories in label
    """
    #Calculating average split half matrix
    avgSplitHalfMat = split_half_analysis(label_var)
    #Plotting
    plt.figure(figsize=(7,7))
    plt.imshow(avgSplitHalfMat,cmap='plasma',vmin=0,vmax=1)
    plt.colorbar()
    plt.xticks(np.unique(data[label_var]),categoryNames,fontsize=14)
    plt.yticks(np.unique(data[label_var]),categoryNames,fontsize=14)
    plt.xlabel('First Half',fontsize=18)
    plt.ylabel('Second Half',fontsize=18)
    for i in range(avgSplitHalfMat.shape[0]):
        for j in range(avgSplitHalfMat.shape[1]):
            plt.text(j-.2, i+.025, '%.3f' % (avgSplitHalfMat[i, j]), color='k',fontsize=14)
```

# Conclusions
Logistic regressions revealed that attitudes, norms and self-efficacy around contemplative practice seem to influence a participant’s intentions to engage in these practices. This is further supported by k-means clustering and analyses similar to representational similarity and split-half reliability analyses, which revealed that participants with similar intentions to engage in contemplative practice had similar attitudes, norms and self-efficacies around contemplative practice. Furthermore, the ability of attitudes, norms and self-efficacy to predict intentions is core to the theoretical basis of this study, the Theory of Planned Behavior. Unfortunately, these attitudes, norms and self-efficacies do not fully explain the influence age has on a participant’s intentions. Thus, it appears as though there are other psychological factors related to age that should be explored to fully understand the effect age has on someone’s intentions to engage in contemplative practice.

Education appears to be the only demographic factor which plays a significant role in determining how likely a person is to ever engage in contemplative practice. However, the cross validated models trained and tested to predict whether a participant will ever engage in contemplative practice using their highest level of education were not very accurate. This is a good result to find, as it indicates that the influence education, and all other demographic factors analyzed, on engagement in contemplative practice is fairly weak and can thus potentially be addressed through interventions.

It does not appear as though a linear mixed effects model controlling for education or age would be particularly appropriate in predicting a participant’s summed self-reported engagement from their intentions. Most other studies applying the Theory of Planned Behavior find that intentions are a very good predictor of behavior. Thus, the fact that predictions of summed self-reported engagement did not improve much when using only participants of particular age or education groups suggests that there may be issues with how intentions or behavior were measured in the study. For example, the summed self-reported engagement in contemplative practice may not be compatible with the measurement of intention used in the study. The intention measure in this study asked participants to indicate how likely it was for them to engage in five minutes of contemplative practice every day for a week, not how often they anticipated engaging in contemplative practice. The intention measure used in this study may be more predictive of a dichotomous variable indicating whether someone engaged in at least five minutes of contemplative practice every day during the week which their behavior was tracked. As only 58 participants (29.74%) completed all of the self-reports they received during the intervention, analysis of this dichotomous variable would likely have been underpowered and was thus not pursued. Instead, this study would recommend that future research use a participant’s intention to engage in contemplative practice once a week to predict whether the participant ever engaged in contemplative practice over the course of a week.

Lastly, there may be some very weak similarities in attitudes, norms and self-efficacies for participants of similar summed self-reported engagement. Specifically, analyses similar to representational similarity and split-half reliability analyses found that participants with the highest and lowest self-reported engagement seemed to have more strongly correlated attitudes, norms and self-efficacies. While this potential clustering was not easily observable using t-SNE, a k-means analysis found marginally significant differences in the summed self-reported engagement between predicted cluster groups. This clustering is not surprising given that intentions and the summed self-reported engagement were weakly positively correlated with each other. However, the attitudes, norms and self-efficacies do not group enough by level of engagement for these measures to be able to predict a participant’s summed self-reported engagement well, as found in linear regression analysis prior to the current set of analyses. Unfortunately, this means that my analyses were not able to explain why participants may have engaged in less contemplative practice during the intervention.
